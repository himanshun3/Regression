```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df1=pd.read_csv('character-deaths.csv')
x=df1['Death Year']
y=df1['Death Chapter']
plt.scatter(x,y,label='data',color='red',marker='+')
sns.regplot(x='Death Year',y='Death Chapter',data=df1,label='Reg1',color='blue',order=1,scatter=None)
sns.regplot(x='Death Year',y='Death Chapter',data=df1,label='Reg2',color='green',order=2,scatter=None)
plt.legend(loc='upper right')
plt.show()
```
